## Description:
#### This module simply plots the data from the ANSYS depth analysis.
#### In addition, it corrects the x-axis depth values to match anatomical predictions. Model 0 mm depth was defined as 0.75 mm deep.
- Input: Voltage data from ANSYS: .csv output 
- Output: .eps file containing the plotted data for Supp. fig 4.

# Setup

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Process data

```python
data=pd.read_csv('/home/jovyan/ansys_data/depth-compare-final.csv')

v_disc_500 = data.v_disc_500
v_disc_850 = data.v_disc_850
v_tetrode_500 = data.v_tetrode_500
v_tetrode_850 = data.v_tetrode_850
depth = (data.e_z)*-1 + 0.65
```

# Plot

```python
# First, clear old plot if one exists
plt.clf()
# Now, create figure & add plots to it
fig, ax1 = plt.subplots(figsize=[8.5, 4], dpi=300)
ax2 = ax1.twinx()
ax2.plot(depth, v_disc_500 / v_tetrode_850, color='pink', linestyle='dotted')
ax2.plot(depth, v_disc_850 / v_tetrode_850, color='purple', linestyle='dotted')
ax2.set_ylim(1,4)
ax1.plot(depth, v_disc_500)
ax1.plot(depth, v_disc_850)
ax1.plot(depth, v_tetrode_500)
ax1.plot(depth, v_tetrode_850)
fig.legend(["DISC 500um", "DISC 850um", "Tetrode 500um", "Tetrode 850um", "DISC 850um/Tetrode 500um", "DISC 850um/Tetrode 850um"])

plt.xlabel("e_z [um]")

# ax1.ylabel("Voltage Ratio (DISC/tetrode)")
fig.savefig('/home/jovyan/ansys_data/images/depth-compare.eps', format='eps', dpi=500)
plt.show()
```
