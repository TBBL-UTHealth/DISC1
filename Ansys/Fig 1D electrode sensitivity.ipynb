## Description:
#### This module performs simple calculations for computing normalized voltage and plotting the result shown in Inset D for Figure 1.
#### Note: For electrode radius >400, the diameter calculation around the cylinder is not valid because a ring forms.
#### The ansys data is averaged with gaps of +/- 0.1 & 0.2 microns to smooth out the irregularities in the mesh.
- Input: Voltage data from ANSYS: .csv output 
- Output: .eps file containing the plotted data for Fig. 1D

# Setup

```python
import pandas as pd
# import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
data=pd.read_csv('/home/jovyan/ansys_data/electrode-size-sensitivity-final-avg.csv')
```

# Process data

```python
norm = data.v_front[0]
e_sensitivity = data.v_front / norm
e_diameter = 410 * (2*np.arcsin(data.r_e/410))
```

# Plot data

```python
# First, clear old plot if one exists
plt.clf()
# Now, create figure & add plots to it
plt.figure(figsize=[8.5, 2], dpi=500)
plt.plot(e_diameter, e_sensitivity)
plt.legend(["Electrode Sensitivity"])

plt.xlim(10, 1300) 
plt.xscale("linear")
plt.xlabel("r_e [um]")
plt.yscale("linear")
plt.ylabel("Voltage Ratio (normalized)")
plt.savefig('/home/jovyan/ansys_data/images/e-sensitivity-final-avg.eps', format='eps', dpi=500)
plt.show()
```
