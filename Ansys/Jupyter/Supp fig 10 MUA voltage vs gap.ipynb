# Settings

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data_20_um_src = pd.read_csv('/home/jovyan/ansys_data/mua-6um-filter.csv')
data_100_um_src = pd.read_csv('/home/jovyan/ansys_data/mua-100um-filter.csv')
data_200_um_src = pd.read_csv('/home/jovyan/ansys_data/mua-200um-filter.csv')
```

# Process data

```python
v_65 = data_20_um_src[data_20_um_src['r_l'] == 32.5]
v_120 = data_20_um_src[data_20_um_src['r_l'] == 60]
v_800 = data_20_um_src[data_20_um_src['r_l'] == 400]
v_1200 = data_20_um_src[data_20_um_src['r_l'] == 600]
v_2000 = data_20_um_src[data_20_um_src['r_l'] == 1000]

v100_65 = data_100_um_src[data_100_um_src['r_l'] == 32.5]
v100_120 = data_100_um_src[data_100_um_src['r_l'] == 60]
v100_800 = data_100_um_src[data_100_um_src['r_l'] == 400]
v100_1200 = data_100_um_src[data_100_um_src['r_l'] == 600]
v100_2000 = data_100_um_src[data_100_um_src['r_l'] == 1000]

v200_65 = data_200_um_src[data_200_um_src['r_l'] == 32.5]
v200_120 = data_200_um_src[data_200_um_src['r_l'] == 60]
v200_800 = data_200_um_src[data_200_um_src['r_l'] == 400]
v200_1200 = data_200_um_src[data_200_um_src['r_l'] == 600]
v200_2000 = data_200_um_src[data_200_um_src['r_l'] == 1000]
```

# Plot data

```python
# First, clear old plot if one exists
plt.clf()
# Now, create figure & add plots to it
plt.figure(figsize=[8.5, 12], dpi=500)
plt.plot(v_65['gap'], v_65['v_front'])
plt.plot(v_120['gap'], v_120['v_front'])
plt.plot(v_800['gap'], v_800['v_front'])
plt.plot(v_1200['gap'], v_1200['v_front'])
plt.plot(v_2000['gap'], v_2000['v_front'])

plt.plot(v100_65['gap'], v100_65['v_front'], linestyle='dashed')
plt.plot(v100_120['gap'], v100_120['v_front'], linestyle='dashed')
plt.plot(v100_800['gap'], v100_800['v_front'], linestyle='dashed')
plt.plot(v100_1200['gap'], v100_1200['v_front'], linestyle='dashed')
plt.plot(v100_2000['gap'], v100_2000['v_front'], linestyle='dashed')

plt.plot(v200_65['gap'], v200_65['v_front'], linestyle='dotted')
plt.plot(v200_120['gap'], v200_120['v_front'], linestyle='dotted')
plt.plot(v200_800['gap'], v200_800['v_front'], linestyle='dotted')
plt.plot(v200_1200['gap'], v200_1200['v_front'], linestyle='dotted')
plt.plot(v200_2000['gap'], v200_2000['v_front'], linestyle='dotted')

plt.legend(["D_L=65", "D_L=120", "D_L=800", "D_L=1200", "D_L=2000", "D_L=65", "D_L=120", "D_L=800", "D_L=1200", "D_L=2000", "D_L=65", "D_L=120", "D_L=800", "D_L=1200", "D_L=2000"])

plt.xscale("linear")
plt.xlabel("gap [um]")
plt.ylim(0,400)
plt.yscale("linear")
plt.ylabel("Voltage [uV]")
plt.savefig('/home/jovyan/ansys_data/images/mua-analysis.eps', format='eps', dpi=500)
plt.show()
```
