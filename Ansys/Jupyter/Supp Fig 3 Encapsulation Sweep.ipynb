# Settings

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data=pd.read_csv('/home/jovyan/ansys_data/encapsulation-02cond-avg-with-05.csv')
```

# Process data

```python
v_05 = data[data['conductivity'] == 0.05]
v_10 = data[data['conductivity'] == 0.1]
v_15 = data[data['conductivity'] == 0.15]
v_20 = data[data['conductivity'] == 0.2]
v_32 = data[data['conductivity'] == 0.32]
v_40 = data[data['conductivity'] == 0.4]
```

# Plot data

```python
# First, clear old plot if one exists
plt.clf()
# Now, create figure & add plots to it
plt.figure(figsize=[8.5, 2], dpi=500)
plt.plot(v_05['r_enc'], v_05['v_front'])
plt.plot(v_10['r_enc'], v_10['v_front'])
plt.plot(v_15['r_enc'], v_15['v_front'])
plt.plot(v_20['r_enc'], v_20['v_front'])
plt.plot(v_32['r_enc'], v_32['v_front'])
plt.plot(v_40['r_enc'], v_40['v_front'])
plt.legend(["Sigma=0.05", "Sigma=0.1", "Sigma=0.15", "Sigma=0.2", "Sigma=0.32", "Sigma=0.40"])

# plt.xlim(20, 200) 
plt.xscale("linear")
plt.xlabel("r_e [um]")
plt.ylim(60, 90)
plt.yscale("linear")
plt.ylabel("Voltage")
plt.savefig('/home/jovyan/ansys_data/images/encapsulation-02cond-avg-with-05.eps', format='eps', dpi=500)
# plt.savefig('/home/jovyan/ansys_data/images/e-sensitivity500.jpg', format='jpeg', dpi=500)
# plt.savefig('/home/jovyan/ansys_data/images/e-sensitivity500.png', format='png', dpi=500)
plt.show()
```
